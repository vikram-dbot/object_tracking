```python
import os
import json
from ultralytics import YOLO
import supervision as sv
```

```python
model = YOLO('best (1).pt')
```

```python
input_path = "C:/Users/Welcome/phases_detection/output/video24/"
output_path = "C:/Users/Welcome/phases_detection/output/video24/output/"
os.makedirs(output_path, exist_ok=True)
for file in os.listdir(input_path):
    if file.lower().endswith('.jpg'):
        image_path = os.path.join(input_path, file)
        print(file)

        # Run YOLOv8 inference
        results = model(image_path, save=True)

        # Get JSON string from results
        result_json_str = results[0].tojson()

        # Convert JSON string to Python object before saving
        try:
            result_json_data = json.loads(result_json_str)

            # Save as proper JSON file
            output_file = os.path.join(output_path, f"{os.path.splitext(file)[0]}.json")
            with open(output_file, 'w') as f:
                json.dump(result_json_data, f, indent=2)

            print(f"Saved JSON for {file}")
        except json.JSONDecodeError as e:
            print(f"Error decoding JSON for {file}: {e}")
```

```python
def parse_annotation(annotation_data):
    with open(annotation_data,'r') as f:
        content = f.read()
        data = json.loads(content)
    bboxes = []
    for detection in data:
        # Safely get the box dictionary from the detection
        box = detection.get("box", {})
        
        # Extract coordinates with type checking (optional)
        try:
            x1 = float(box.get("x1")) if "x1" in box else None
            y1 = float(box.get("y1")) if "y1" in box else None
            x2 = float(box.get("x2")) if "x2" in box else None
            y2 = float(box.get("y2")) if "y2" in box else None
        except (TypeError, ValueError):
            continue
        
        if all(val is not None for val in (x1, y1, x2, y2)):
            bboxes.append((x1, y1, x2, y2))
    
    return bboxes
```

```python
annotation_data = r"output\video24\output\video24_3.json"
box = parse_annotation(annotation_data)
```
